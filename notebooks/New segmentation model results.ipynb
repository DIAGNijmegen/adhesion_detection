### nnU-Net training curves

```python
train_stat_path = Path("../../final output/nnUnet")
train_curves_file = "progress.png"
folds_folders = [f.name for f in train_stat_path.iterdir() if f.is_dir()]
folds_folders.sort()
print(folds_folders)
```

```python
for ind, fold in enumerate(folds_folders):
    progress_path = train_stat_path / fold / train_curves_file
    progress_image = io.imread(progress_path)
    
    plt.figure(figsize=(15,20))
    plt.imshow(progress_image)
    plt.axis("off")
    plt.title("Fold {}".format(ind + 1))
    plt.show
```

```python
gifs_path = Path("../../experiments/final_detection_segm_gifs")
vs_path = Path("../../data/cinemri_mha_detection/visualization/cum_vs_warp_contour_norm_avg_rest")
```

# Positive

```python
positive_glob = gifs_path.glob("CM0*")
positive_ids = [file.stem for file in positive_glob]

for positive_id in positive_ids:
    print(positive_id)
    vs_img_path = vs_path / (positive_id + ".png")
    vs_gif_path = gifs_path / (positive_id + ".gif")
    
    vs_img_file = open(vs_img_path, "rb")
    vs_gif_file = open(vs_gif_path, "rb")
    
    wi1 = widgets.Image(value=vs_gif_file.read(), format='png')
    wi2 = widgets.Image(value=vs_img_file.read(), format='png')
    
    sidebyside = widgets.HBox([wi1, wi2])
    ## Finally, show.
    display.display(sidebyside)
    
    vs_img_file.close()
    vs_gif_file.close()
```

# Negative

```python
negative_glob = vs_path.glob("R*")
negative_ids = [file.stem for file in negative_glob]

for negative_id in negative_ids:
    print(negative_id)
    vs_img_path = vs_path / (negative_id + ".png")
    vs_gif_path = gifs_path / (negative_id + ".gif")
    
    vs_img_file = open(vs_img_path, "rb")
    vs_gif_file = open(vs_gif_path, "rb")
    
    wi1 = widgets.Image(value=vs_gif_file.read(), format='png')
    wi2 = widgets.Image(value=vs_img_file.read(), format='png')
    
    sidebyside = widgets.HBox([wi1, wi2])
    ## Finally, show.
    display.display(sidebyside)
    
    vs_img_file.close()
    vs_gif_file.close()
```
